# Visualizing The Gender Gap In College Degrees

In this guided project I have used various methods in matplotlib to visualise the gender gap in college degrees. The dataset that I have used is obtained from Randal Olson, a data scientist at University of Pennsylvania, he has cleaned the data set and made it available on his personal website. You can download the dataset Randal compiled http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv.

# Generating plots for STEM degrees  

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

# Comparing across all degrees

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(16,20))


#Generating plots for STEM degrees in the first column

for i in range(0,18,3):
    stem_index = int(i/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[stem_index]], c=cb_dark_blue, label="Women", linewidth=3)
    ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[stem_index]], c=cb_orange, label="Men", linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[stem_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    if stem_index == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif stem_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')

#Generating plots for lib_arts_degrees in second column

for i in range(1,16,3):
    arts_index = int((i-1)/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[arts_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[arts_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[arts_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if arts_index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')

#Generating plots for other_cats degrees in third column

for i in range(2,20,3):
    other_index = int((i-2)/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[other_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[other_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[other_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    if other_index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif other_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        
plt.show()
        
```
